```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=10,factor = 0.5)

print(X.shape,y.shape)
print(X)
print(y)
```

```python
X,y = make_circles(n_samples=1000,factor = 0.4)  # factor controls the distace between 2 circles
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
split = 700

X_train = X[:split,:]
X_test = X[split:,:]

y_train = y[:split]
y_test = y[split:]

print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

## 1. logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test) * 100
```

## 2. KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train,y_train)
knn.score(X_test,y_test) *100
```

```python
X,y = make_circles(n_samples=50,factor = 0.8)  # factor controls the distace between 2 circles
plt.scatter(X[:,0],X[:,1])
plt.show() # This below graph now will starts creating a problem
```

```python
split = int(0.7 * X.shape[0])

X_train = X[:split,:]
X_test = X[split:,:]

y_train = y[:split]
y_test = y[split:]

print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train,y_train)
knn.score(X_test,y_test) *100
```

## 3. Neural Networks

```python
X,y = make_circles(n_samples=500,factor = 0.4)  # factor controls the distace between 2 circles
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
split = int(0.7 * X.shape[0])

X_train = X[:split,:]
X_test = X[split:,:]

y_train = y[:split]
y_test = y[split:]

print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(solver='lbfgs',
                    alpha = 1e-5,
                   hidden_layer_sizes=(6),
                   random_state=100,#seed
                   activation = 'logistic',
                   max_iter =200)

mlp.fit(X_train,y_train)
mlp.score(X_test,y_test) * 100
```
