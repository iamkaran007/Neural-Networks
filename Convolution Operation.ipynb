```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
im = cv2.imread('img/landscape.jpg')
img = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

print(img.shape)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
K = np.array([
    [0,-1,0],
    [-1,4,-1],
    [0,-1,0]
])

print(K.shape)
```

```python
out_r = img.shape[0] - K.shape[0] +1
out_c = img.shape[1] - K.shape[1] +1

new_img = np.zeros((out_r,out_c))
print(new_img.shape)

```

```python
for r in range(out_r):
    for c  in range(out_c):
        image_patch = img[r:r+K.shape[0],c:c+K.shape[1]]
        prod = image_patch * K
        prod = prod.sum()
        new_img[r,c] = prod if prod > 0 else 0
```

```python
plt.imshow(new_img,cmap='gray')
plt.show()
```
